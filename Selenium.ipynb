```python
import requests
#import html5lib
from bs4 import BeautifulSoup

URL = "https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&as-pos=1&as-type=RECENT&suggestionId=laptop%7CLaptops&requestId=29b734b0-a716-4f3d-8ce4-83c62eca7ce1&as-searchtext=laptop"
r = requests.get(URL)

soup = BeautifulSoup(r.content, 'html') # If this line causes an error, run 'pip install html5lib' or install html5lib
print(soup.prettify())
```
