```python
#ROCK PAPER SISSORS
from random import choice
win=[('r','s'),('p','r'),('s','p')]
name=input("Enter your name: ")
print("*"*80)
print("||Welcome to Rock Paper Sissors||".center(80))
print("*"*80)
while True:
    com=choice(['r','p','s'])
    player=input("Enter your choice between 'r','p','s'   ")
    if (player,com) in win:
        print(f"{name} WINSSSS")
    elif  player==com:
        print("MATCH DRAWWW!!!")
    else:
        print(f"COMPUTER WINSSS")
    ch=input("Do you want to play again(y/n) -->").lower().strip()
    if ch=="y" or ch=="yes":
        continue
    else:
        print("Thankyou for playing!!!")
        break
```

```python
from random import randint
com=randint(1,10)
c=0
name=input("Enter your name: ")
while True:
    while c<=5:
        player=int(input("Enter your number: "))
        if player == com:
            print(f"{name} WINSSS")
            break
        elif com>player:
            print("Soch badi karrr")
        else:
            print("Kam socha karrr")
        c+=1
    else:
        print(f"COMPUTERR WINSS!! by {com}")
    ch=input("Do you want to play again(y/n) -->").lower().strip()
    if ch=="y" or ch=="yes":
        continue
    else:
        print("Thankyou for playing!!!")
        break
```
