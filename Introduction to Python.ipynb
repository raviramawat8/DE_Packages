```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,360)
```

```python
sin = np.sin(x)
cos = np.cos(x)
```

```python
plt.figure(figsize=(15,5))
plt.plot(sin,'r-.',label='SIN')
plt.plot(cos,'g-.',label="COS")
plt.plot([0,360],[0,0],'k')
plt.legend()
plt.show()
```
